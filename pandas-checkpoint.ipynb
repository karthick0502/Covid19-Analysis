```python
df
```

```python
df.dropna().head()
```

```python
df.dropna(axis='index',how='any')

#axis parameter can have passing as 'index' and 'columns'
```

```python
df.dropna(axis='index',how='all')
#how='all'- can be drop when that index having Null for all the cloumns
```

```python
df.dropna(axis='columns',how='all')
#this method can be drop when single columns have none values
```

```python
df.dropna(axis='columns',how='any')
```

```python
#subset can be used to drop the nulll values depends on particular columns
```

```python
df.dropna(axis='index',how='all',subset=['country','language'])
#when we use how parameter as all then it is drop when all of the subset values are Null((it is like AND gate)
```

```python
df.dropna(axis='index',how='any',subset=['country','language'])
#when we use how parameter as any then it is drop when any one of the subset values are Null (it is like OR gate)
```
