### part5- Filling area on line plots

```python
#in line plot- we can use plt.fill_between() mehthod to fill the the plots where exact condition matches
```

```python
cdf=pd.read_csv('k:\project work\COVID-19_Daily_Cases__Deaths__and_Hospitalizations.csv')
```

```python
cdf=cdf.dropna(axis='index',how='any')
```

```python
cdf
```

```python
filt=(cdf['Date']>='03/01/2020') & (cdf['Date']<='03/30/2020')
```

```python
x=cdf.loc[filt,'Date']
y=cdf.loc[filt,'Cases - Age 18-29']
y1=cdf.loc[filt,'Cases - Age 30-39']
```

```python
plt.plot(x,y,label='ages 18-29')
plt.plot(x,y1,label='ages 30-39')
plt.fill_between(x,y)
plt.legend()
plt.title('covid cases graph')
plt.xlabel('dates')
plt.ylabel('covid cases age 18-29')
plt.tight_layout()
```

```python
plt.plot(x,y,label='ages 18-29')
plt.plot(x,y1,label='ages 30-39')
plt.fill_between(x,y1,alpha=0.25) #aplha can use the fil with dim shades so that we can easily look up on those two edges
plt.legend()
plt.title('covid cases graph')
plt.xlabel('dates')
plt.ylabel('covid cases age 18-29')
plt.tight_layout()
```

```python
ym=cdf.loc[filt,'Cases - Age 18-29'].median()
#y1m=cdf.loc[filt,'Cases - Age 30-39']
```

```python
plt.plot(x,y,label='ages 18-29')
plt.plot(x,y1,label='ages 30-39')
plt.fill_between(x,y,ym,alpha=0.25) #when we want to fill the shades below and above median value then we can give that median value in third parameter 
                                   #in fill_between()
plt.legend()
plt.title('covid cases graph')
plt.xlabel('dates')
plt.ylabel('covid cases age 18-29')
plt.tight_layout()
```

```python
plt.plot(x,y,label='ages 18-29')
plt.plot(x,y1,label='ages 30-39')
plt.fill_between(x,y,ym,where=(y<ym),interpolate=True,alpha=0.25) #mark the shades only condition applied in where clause
plt.legend()
plt.title('covid cases graph')
plt.xlabel('dates')
plt.ylabel('covid cases age 18-29')
plt.tight_layout()
```

```python
plt.plot(x,y,label='ages 18-29')
plt.plot(x,y1,label='ages 30-39')
plt.fill_between(x,y,ym,where=(y<ym),interpolate=True,alpha=0.25) #mark the shades only condition applied in where clause
plt.fill_between(x,y,ym,where=(y>=ym),interpolate=True,alpha=0.25) #we can apply diff condition using multiple fill_between()
plt.legend()
plt.title('covid cases graph')
plt.xlabel('dates')
plt.ylabel('covid cases age 18-29')
plt.tight_layout()
```

```python
plt.plot(x,y,label='ages 18-29')
plt.plot(x,y1,label='ages 30-39')
plt.fill_between(x,y,ym,where=(y<ym),interpolate=True,color='green',alpha=0.25) #mark the shades only condition applied in where clause
plt.fill_between(x,y,ym,where=(y>=ym),interpolate=True,color='red',alpha=0.25) #we can apply diff condition using multiple fill_between()
plt.legend()
plt.title('covid cases graph')
plt.xlabel('dates')
plt.ylabel('covid cases age 18-29')
plt.tight_layout()
```

```python
plt.plot(x,y,label='ages 18-29')
plt.plot(x,y1,label='ages 30-39')
plt.fill_between(x,y,y1,where=(y<y1),interpolate=True,color='green',alpha=0.25) #mark the shades only condition applied in where clause
plt.fill_between(x,y,y1,where=(y>=y1),interpolate=True,color='red',alpha=0.25) #we can apply diff condition using multiple fill_between()
plt.legend()
plt.title('covid cases graph')
plt.xlabel('dates')
plt.ylabel('covid cases age 18-29')
plt.tight_layout()
```

```python
plt.plot(x,y,label='ages 18-29')
plt.plot(x,y1,label='ages 30-39')
plt.fill_between(x,y,y1,where=(y<y1),interpolate=True,color='green',alpha=0.25,label='below 18-29 when compare than 30-39') 
plt.fill_between(x,y,y1,where=(y>=y1),interpolate=True,color='red',alpha=0.25,label='above 18-29 when compare than 30-39') 
plt.legend()
plt.title('covid cases graph')
plt.xlabel('dates')
plt.ylabel('covid cases age 18-29')
plt.tight_layout()
```

### we can use unpacking method to split the axisplots

```python
fig,ax=plt.subplots(nrows=2,ncols=1)
print(ax)

#here, tow axessubplot are existing within single list
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1)
print(ax1)
print(ax2)

#here, two axessubplots are seperately existing
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1)

ax1.plot(xs,ys,label='SQL',color='red') #in first axis we are plotting sql alone only
ax2.plot(xp,yp,label='Python') #in second axis we are plotting python and dart together
ax2.plot(xd,yd,label='Dart')


ax1.legend()
ax1.set_title('salaries of each lanugage in india')
#ax1.set_xlabel('ages')
ax1.set_ylabel('salaries')


ax2.legend()
#ax2.set_title('salary in Python & Dart')
ax2.set_xlabel('ages')
ax2.set_ylabel('salaries')


```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,sharex=True) #sharex can be use to sharing the x axis values(it can applicable in y axis)

ax1.plot(xs,ys,label='SQL',color='red') 
ax2.plot(xp,yp,label='Python') 
ax2.plot(xd,yd,label='Dart')


ax1.legend()
ax1.set_title('salaries of each lanugage in india')
#ax1.set_xlabel('ages')
ax1.set_ylabel('salaries')


ax2.legend()
#ax2.set_title('salary in Python & Dart')
ax2.set_xlabel('ages')
ax2.set_ylabel('salaries')
```

```python
fig1,ax1=plt.subplots()
fig2,ax2=plt.subplots() #we can seperate the figures as well

ax1.plot(xs,ys,label='SQL',color='red') 
ax2.plot(xp,yp,label='Python') 
ax2.plot(xd,yd,label='Dart')


ax1.legend()
ax1.set_title('salaries of each lanugage in india')
ax1.set_xlabel('ages')
ax1.set_ylabel('salaries')


ax2.legend()
ax2.set_title('salary in Python & Dart')
ax2.set_xlabel('ages')
ax2.set_ylabel('salaries')
```
